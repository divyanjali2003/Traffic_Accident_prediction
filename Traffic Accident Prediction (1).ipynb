```python
import pandas as pd
```

```python
df=pd.read_csv("kerala_accident_data_with_datetime.csv")
```

```python
df
```

```python
features=['Latitude','Longitude','Severity','Weather_Condition','Road_Type','Vehicle_Type']
```

```python
df1=df[features]
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label1=LabelEncoder()
label2=LabelEncoder()
label3=LabelEncoder()
label4=LabelEncoder()
```

```python
df1["Weather_Condition"]=label1.fit_transform(df["Weather_Condition"])
df1["Vehicle_Type"]=label2.fit_transform(df["Vehicle_Type"])
df1["Severity"]=label3.fit_transform(df["Severity"])
df1["Road_Type"]=label4.fit_transform(df["Road_Type"])
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
x=df1[features]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,11),wcss,marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('Intertia')
plt.show()
```

```python
optimal_k = 6
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df1['Cluster'] = kmeans.fit_predict(x_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df1['Longitude'], df1['Latitude'], c=df1['Cluster'], cmap='viridis', s=5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title(f'Traffic Accident Hotspots - {optimal_k} Clusters')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score 
```

```python
silhouette_avg = silhouette_score(x_scaled, df1['Cluster'])
print(f'Silhouette Score: {silhouette_avg}')
```

```python
import pickle
with open('kmeans_model.pkl', 'wb') as model_file:
    pickle.dump(kmeans, model_file)
with open('kmeans_scaler.pkl', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)
if 'label_encoder' in locals():  
    with open('kmeans_label.pkl', 'wb') as label_file:
        pickle.dump(label_encoder, label_file)
print("Model, scaler, and label encoder (if used) saved successfully.")
```
